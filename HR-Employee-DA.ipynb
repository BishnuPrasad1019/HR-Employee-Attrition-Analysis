Fill value

```python
# Numerical columns → fill with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Categorical columns → fill with mode
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean())
```

Replaceing numerical values to categorical values

```python
df['RelationshipSatisfaction'].unique()
df['StockOptionLevel'].unique()
df['WorkLifeBalance'].unique()
```

```python
df[['RelationshipSatisfaction','WorkLifeBalance']]=df[['RelationshipSatisfaction','WorkLifeBalance']].replace({1:'Great',2:'Good',3:'Average',4:'Not Good'})

df['StockOptionLevel']=df['StockOptionLevel'].replace({0:'Not Investing',1:'Recently Start Investing',2:'Average Investing',3:'Good Investing'})
```

```python
df['Gender']=df['Gender'].str.lower()
df['Department']=df['Department'].str.strip()
```

Groupby function 

```python
df['MonthlyIncome']=df['MonthlyIncome'].replace({'MonthlyIncome':'Salary'})
```

```python
df.groupby('Department')['MonthlyIncome'].sum()
```

```python
df.groupby('Department')['MonthlyIncome'].agg(['count','min','max','mean','sum'])
```

```python
df.groupby(['Department','Gender'])['MonthlyIncome'].mean()
```

Outliers

```python
Q1 = df['MonthlyIncome'].quantile(0.25)
Q3 = df['MonthlyIncome'].quantile(0.75)
IQR = Q3 - Q1
```

```python
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['MonthlyIncome'] < lower_limit) | (df['MonthlyIncome'] > upper_limit)]
outliers
```

```python
df_no_outliers = df[(df['MonthlyIncome'] >= lower_limit) & (df['MonthlyIncome'] <= upper_limit)]
```

Visualization

```python
dep_income = df.groupby('Department')['MonthlyIncome'].sum().to_frame()

sns.heatmap(dep_income, annot=True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

dep_income = df.groupby(['Department', 'MonthlyIncome'])['MonthlyRate'].sum().reset_index()
pivot_table = dep_income.pivot(
    index='Department',
    columns='MonthlyIncome',
    values='MonthlyRate',
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, fmt=".0f")
plt.title("Department vs Monthly Income - Salary Heatmap")
plt.show()
```

```python
# Separate column types
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(exclude='number').columns

print("Numeric Columns:", num_cols)
print("Categorical Columns:", cat_cols)
```

```python
for col in cat_cols:
    plt.figure(figsize=(5, 5))
    df[col].value_counts().plot(kind='bar')
    plt.title(f"Count of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
for col in num_cols:
    plt.figure()
    plt.hist(df[col].dropna(), bins=15)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
for col in num_cols:
    plt.figure(figsize=(5, 5))
    plt.boxplot(df[col].dropna())
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.show()
```

```python
grouped_income = df.groupby('Department')['MonthlyIncome'].mean()
plt.figure(figsize=(5, 4))
grouped_income.plot(kind='bar')
plt.title("Average Monthly Income by Department")
plt.xlabel("Department")
plt.ylabel("Average Monthly Income")
plt.show()
```

```python
attrition_count = df['Attrition'].value_counts()
plt.figure(figsize=(5, 5))
attrition_count.plot(kind='bar')
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Number of Employees")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(df['TotalWorkingYears'], df['MonthlyIncome'])
plt.title("Monthly Income vs Total Working Years")
plt.xlabel("Total Working Years")
plt.ylabel("Monthly Income")
plt.show()
```

```python
# GROUP-WISE BOXPLOT (ADVANCED – INTERVIEW LEVEL)

plt.figure(figsize=(5, 10))
df.boxplot(column='MonthlyIncome', by='Department')
plt.title("Monthly Income by Department")
plt.suptitle("")
plt.xlabel("Department")
plt.ylabel("Monthly Income")
plt.show()
```
